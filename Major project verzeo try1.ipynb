```python
from bs4 import BeautifulSoup as bs
import string
from nltk.corpus import stopwords 
from nltk.stem import WordNetLemmatizer as wnl
from nltk.tokenize import RegexpTokenizer as ret
from nltk.stem.porter import PorterStemmer as ps

```

def rem_punc(text) :
    puncless="".join([c for c in text not in string.punctuation])
    return puncless
wm1['Review Text']=wm1['Review Text'].apply(lambda x : rem_punc())
wm1['Review Text'].head(20)

```python
# Remove punctuatuions and lower words.
tokenizer = ret(r'\w+')
wm1['Review Text']= wm1['Review Text'].apply(lambda x : tokenizer.tokenize(x.lower()))
wm1['Review Text'].head(10)
```

```python
#remove stopwords
def rem_stopwords(text) :
    word=[w for w in text if w not in stopwords.words('english')]
    return word
wm1['Review Text']= wm1['Review Text'].apply(lambda x : rem_stopwords(x))
wm1['Review Text'].head(10)
```

```python
lemmatizer=wnl()
def lemm(text) : 
    lt=[lemmatizer.lemmatize(i) for i in text]
    return lt
wm1['Review Text']=wm1['Review Text'].apply(lambda x : lemm(x))
wm1['Review Text'].head(10)
```

```python
# Remove numbers
def rem_digit(text) :
    output = ' '.join(c for c in text if not c.isdigit())
    return output
wm1['Review Text']=wm1['Review Text'].apply(lambda x : rem_digit(x))
wm1['Review Text'].head(10)
```

```python
wmc=wm1.copy()
wmc.head(10)
```

```python
#tfidf vectoriziation output (a,b) c = a:=no of reviews, b:=word vector index c:= value of tfidf
from sklearn.feature_extraction.text import TfidfVectorizer as tfidfv
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
tfidf = tfidfv(max_features=1000,ngram_range=(1,2),analyzer='word')
wmct= tfidf.fit_transform(wmc['Review Text'])
print(wmct)
```

```python
#bag of words

bow = CountVectorizer(max_features=1000, lowercase=True, ngram_range=(1,1),analyzer = "word")
wmcb = bow.fit_transform(wmc['Review Text'])
print(wmcb)
```
