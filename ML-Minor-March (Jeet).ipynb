```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from datetime import date as date
from datetime import datetime as dt
```

```python
apps = pd.read_csv(r'H:\machine  learning internship\verzeo\appstore_games (5).csv')

apps
```

```python
apps.drop(['URL','ID','Subtitle','Icon URL','Description','Developer','Primary Genre'],axis=1,inplace=True)
apps.head()
```

```python
apps.rename(columns={'Average User Rating':'Average_User_Rating','User Rating Count':'User_Rating_Count','In-app Purchases':'In_app_Purchases','Age Rating':'Age_Rating','Original Release Date':'Original_Release_Date','Current Version Release Date':'Current_Version_Release_Date'},inplace=True)
apps.head()
```

```python

apps.dtypes
apps['Current_Version_Release_Date'] = pd.to_datetime(apps['Current_Version_Release_Date'])
apps['Original_Release_Date'] = pd.to_datetime(apps['Original_Release_Date'])
apps.dtypes
apps['Upgrade']= (apps['Current_Version_Release_Date'] - apps['Original_Release_Date']).dt.days
apps
```

```python
apps.dropna(inplace=True)

apps
```

```python
lst=[]
lst=apps[(apps.User_Rating_Count<200 )&(apps.Upgrade<180)]

apps.drop(lst.index,axis=0,inplace=True)
apps
```

```python
apps.Genres=apps.Genres.astype(str)
lst6=apps.Genres.str.split(",",expand =True)
apps['Genres1']=lst6[0]
apps['Genres2']=lst6[1]
apps['Genres3']=lst6[2]
apps['Genres4']=lst6[3]




apps.Genres1=apps.Genres1.str.replace('Entertainment','NaN')
apps.Genres1=apps.Genres1.str.replace('Strategy','NaN')
apps.Genres1=apps.Genres1.str.replace('Games','NaN')
apps.Genres1=apps.Genres1.str.replace('None','NaN')
apps.Genres2=apps.Genres2.str.replace('Entertainment','NaN')
apps.Genres2=apps.Genres2.str.replace('Strategy','NaN')
apps.Genres2=apps.Genres2.str.replace('Games','NaN')
apps.Genres2=apps.Genres2.str.replace('None','NaN')
apps.Genres3=apps.Genres3.str.replace('Entertainment','NaN')
apps.Genres3=apps.Genres3.str.replace('Strategy','NaN')
apps.Genres3=apps.Genres3.str.replace('Games','NaN')
apps.Genres3=apps.Genres3.str.replace('None','NaN')
apps.Genres4=apps.Genres4.str.replace('Entertainment','NaN')
apps.Genres4=apps.Genres4.str.replace('Strategy','NaN')
apps.Genres4=apps.Genres4.str.replace('Games','None')
apps.Genres4=apps.Genres4.str.replace('None','NaN')
apps.Genres=apps.Genres.astype(str)
apps['NGenres']=apps[['Genres1', 'Genres2', 'Genres3', 'Genres4']].apply(lambda x: ', '.join(x.dropna()), axis = 1)
apps.NGenres=apps.NGenres.str.replace('NaN,','')
apps.NGenres=apps.NGenres.str.replace(', NaN,','')
apps.NGenres=apps.NGenres.str.replace('NaN','')
apps.NGenres=apps.NGenres.str.replace(' ','')

apps
```

```python
apps.NGenres=apps.NGenres.str.replace('Board','Puzzle',regex=True)
apps.NGenres=apps.NGenres.str.replace('Role Playing','Adventure',regex=True)
apps.NGenres=apps.NGenres.str.replace('Role','Adventure',regex=True)
apps.NGenres=apps.NGenres.str.replace('Education','Family',regex=True)
apps
```

```python
apps.Genres=apps.Genres.str.rstrip(',')
apps.NGenres=apps.NGenres.str.rstrip(',')
apps
```

```python
lst8=apps[apps.NGenres=='']
apps.drop(lst8.index,axis=0,inplace=True)
apps
```

```python

apps.groupby('NGenres').Average_User_Rating.mean().plot(kind='bar')
lst9=apps.groupby('NGenres').Average_User_Rating.mean()
BestGenres=lst9.index[lst9 == lst9.max()]
BestGenres    

#Best Genre can be any of them
```

```python
apps.groupby('Size').Average_User_Rating.mean().plot(kind='bar')
lst4=apps.groupby('Size').Average_User_Rating.mean()

BestSize=lst4.index[lst4== lst4.max()]
BestSize

#Best size can be any of them(only size)
```

```python
apps.groupby('Upgrade').Average_User_Rating.mean().plot(kind = 'bar')
lst10=apps.groupby('Upgrade').Average_User_Rating.mean()
BestUpgrade=lst10.index[lst10== lst10.max()]
BestUpgrade  #best upgrade can be any of these.(without Genre)
```

```python
apps.groupby(['NGenres','Price','In_app_Purchases']).Average_User_Rating.mean().plot(kind='hist')
lst11=apps.groupby(['NGenres','Price','In_app_Purchases']).Average_User_Rating.mean()

BestCost=lst11.index[lst11==lst11.max()]
BestCost
#best cost with genre grouped.
```

```python
apps.groupby(['Price','In_app_Purchases']).Average_User_Rating.mean().plot(kind= 'hist')
lst17= apps.groupby(['Price','In_app_Purchases']).Average_User_Rating.mean()
BestCost2=lst17.index[lst17==lst17.max()]
BestCost2
```

```python
apps.groupby('Age_Rating').Average_User_Rating.mean().plot(kind= 'bar')
lst12= apps.groupby('Age_Rating').Average_User_Rating.mean()
BestAge=lst12.index[lst12==lst12.max()]
BestAge
#best age for >4
```

```python

apps.groupby(['NGenres','Size','Upgrade','Price','Age_Rating']).Average_User_Rating.mean().plot(kind='bar')
lst13=apps.groupby(['NGenres','Size','Upgrade','Price','Age_Rating']).Average_User_Rating.mean()

BestCombo=lst13.index[lst13==lst13.max()]
BestCombo
#all combined
```
